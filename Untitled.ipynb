```python
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))
ax = ax.flatten()
num_cols = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration']

for i, col in enumerate(num_cols):
    ax[i].hist(np.array(df[col]))
    ax[i].legend(col)
    ax[i].set_title(str.capitalize(col))
    fig.suptitle("Histograms")
#     ax[i].set_xlable(col)
#     ax[i].set_ylable("Coutn")
    
    
plt.show()
```

```python
model = KernelDensity(bandwidth=2, kernel = 'gaussian')
model_cylin = KernelDensity(bandwidth=2, kernel = 'gaussian')
model_disp = KernelDensity(bandwidth=0.005, kernel = 'gaussian')
model_hp = KernelDensity(bandwidth=2, kernel = 'gaussian')
model_wt = KernelDensity(bandwidth=2, kernel = 'gaussian')
model_acc = KernelDensity(bandwidth=2, kernel = 'gaussian')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))
ax = axes.flatten()
for i, col in enumerate(num_cols):
    clm = np.array(df[col])
    colmn = np.array(df[col]).reshape(-1,1)
    model.fit(colmn)
    values = np.linspace(colmn.min(), colmn.max(), 100).reshape(-1, 1)
    prob_dens = model.score_samples(values)
    prob_dens = np.exp(prob_dens)
    ax[i].hist(clm, density=True)
    ax[i].plot(values[:], prob_dens, 'red')
plt.show()
#     break
```

```python
qrs = df.describe()['25%':]
print(qrs)
```

```python
out_hp = qrs['horsepower']['75%'] + 1.5 * (qrs['horsepower']['75%'] - qrs['horsepower']['25%'])
print(out_hp)
hp_array = np.array(df['horsepower'])
for i in range(len(hp_array)):
    if int(hp_array[i]) > out_hp:
        hp_array[i] = np.median(hp_array)
print(pd.Series(hp_array).describe())
print(df['horsepower'].describe())
df['horsepower'] = hp_array
print(df['horsepower'].describe())
```

```python
out_acc = qrs['acceleration']['75%'] + 1.5 * (qrs['acceleration']['75%'] - qrs['acceleration']['25%'])
print(out_acc)
acc_array = np.array(df['acceleration'])
for i in range(len(acc_array)):
    if int(acc_array[i]) > out_acc:
        acc_array[i] = np.median(acc_array)
print(pd.Series(acc_array).describe())
print(df['acceleration'].describe())
df['acceleration'] = acc_array
print(df['acceleration'].describe())
```
